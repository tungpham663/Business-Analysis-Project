````python
def generation(post_type, keyword):
    """
    Generate a post based on the post type and keyword.

    Parameters:
        post_type (int): The type of post to generate.
        keyword (str): The keyword to include in the post.

    Returns:
        str: The generated post.
    """
    if post_type == 1:
        return gen_1(keyword)
    elif post_type == 2:
        if check_keyword_in_db(keyword):
            paragraph = search_samples_by_keyword(keyword)
            return gen_2(paragraph, keyword)
        else:
            return gen_2_keyword(keyword)
    elif post_type == 3:
        if graph.node_exists(keyword):
            relationships = graph.get_relationships(keyword)
            return gen_3(keyword, relationships)
        else:
            return gen_3_keyword(keyword)
    elif post_type == 4:
        pass
    elif post_type == 5:
        pass
````
